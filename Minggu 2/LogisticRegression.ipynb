```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from google.colab import files
# Mengunggah file dari komputer
uploaded = files.upload()
#membaca menggunakan Pandas
import pandas as pd
SocialNetworkAds = pd.read_csv('SocialNetworkAds.csv')
print(SocialNetworkAds.head())
```

```python
# Menampilkan informasi tentang DataFrame 'SocialNetworkAds'
SocialNetworkAds.info()
```

```python
# Menghapus kolom 'User ID' dari DataFrame 'SocialNetworkAds'
data = SocialNetworkAds.drop(columns=['User ID'])
```

```python
# Menggunakan get_dummies untuk menghasilkan variabel dummy
data = pd.get_dummies(data)

# Hasilnya adalah DataFrame yang telah diubah dengan variabel dummy
data
```

```python
predictions = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']  # Memilih atribut yang akan digunakan sebagai prediktor

x = data[predictions]  # Atribut (fitur) yang digunakan untuk melakukan prediksi
y = data['Purchased']  # Label yang ingin diprediksi

# 'x' berisi atribut-atribut yang digunakan untuk melakukan prediksi (umur, estimasi gaji, jenis kelamin)
# 'y' berisi label yang ingin diprediksi (dalam kasus ini, pembelian (Purchased))
```

```python
# Melakukan standarisasi dari dataset yang akan diolah
scaler = StandardScaler()
scaler.fit(x) # Menghitung mean dan standar deviasi dari dataset X

scaled_data = scaler.transform(x) # Melakukan penskalaan standar pada dataset X
scaled_data = pd.DataFrame(scaled_data, columns = x.columns) # Mengonversi hasil penskalaan menjadi DataFrame dengan nama kolom yang sesuai
scaled_data.head()
```

```python
# Pembagian dataset menjadi data latihan (train) dan data uji (test)
x_train, x_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
# Membuat model regresi logistik
model = linear_model.LogisticRegression()

# Melatih model menggunakan data pelatihan
model.fit(x_train, y_train)
```

```python
# Mengukur kinerja model pada data uji dan menghitung R-squared
score = model.score(x_test, y_test)
print(f"Nilai R-squared pada data uji: {score:.2f}")
```
