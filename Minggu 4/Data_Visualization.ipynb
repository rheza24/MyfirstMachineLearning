```python
import pandas as pd               # Mengimpor modul pandas untuk manipulasi data dan analisis
import matplotlib.pyplot as plt  # Mengimpor modul pyplot dari matplotlib untuk visualisasi data
import seaborn as sns            # Mengimpor modul seaborn untuk visualisasi data dengan lebih menarik
import ipywidgets as widgets     # Mengimpor modul ipywidgets untuk membuat widget interaktif dalam Jupyter notebook
from IPython.display import display  # Mengimpor fungsi display dari IPython untuk menampilkan output di Jupyter notebook
```

```python
# Menampilkan jumlah nilai unik untuk setiap fitur
features = ['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'price', 'x', 'y', 'z']

for feature in features:
    print(f"Jumlah unique value for {feature}: {df[feature].nunique()}")
```

```python
# Mengelompokkan numerical features
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']

# Mengelompokkan categorical features
categorical_features = ['cut', 'color', 'clarity']

# Mengelompokkan target
target = ['price']

# Menampilkan hasil pengelompokkan
print("Numerical Features:", numerical_features)
print("Categorical Features:", categorical_features)
print("Target:", target)
```

```python
# Membuat bar chart untuk setiap fitur kategorikal
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=feature, data=df, palette='viridis')
    plt.title(f'Bar Chart of {feature}', size=15)
    plt.xlabel(feature, size=12)
    plt.ylabel('Count', size=12)
    plt.show()
```

```python

# Mengimpor dataset diamonds.csv ke dalam dataframe
diamonds_df = pd.read_csv('diamonds.csv')

# List fitur kategorikal
categorical_features = ['cut', 'color', 'clarity']

# Set up figure dan axes
fig, axes = plt.subplots(1, len(categorical_features), figsize=(20, 6))

# Plot pie charts untuk setiap fitur kategorikal
for i, feature in enumerate(categorical_features):
    counts = diamonds_df[feature].value_counts()
    axes[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel', len(counts)))
    axes[i].set_title(f'Distribusi {feature.capitalize()}')

plt.tight_layout()
plt.show()
```

```python
# Set up the figure and axes again
fig, axes = plt.subplots(1, len(categorical_features), figsize=(20, 6))

# Plot pie charts for each categorical feature with dark colors
for i, feature in enumerate(categorical_features):
    counts = diamonds_df[feature].value_counts()
    axes[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('dark', len(counts)))
    axes[i].set_title(f'Distribusi {feature.capitalize()}')

plt.tight_layout()
plt.show()
```

```python
# List of numerical features
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Set up the figure and axes
fig, axes = plt.subplots(len(numerical_features), 1, figsize=(10, 20))

# Plot histograms for each numerical feature
for i, feature in enumerate(numerical_features):
    sns.histplot(data=diamonds_df, x=feature, kde=True, ax=axes[i], color='steelblue', bins=50)
    axes[i].set_title(f'Distribusi {feature.capitalize()}')
    axes[i].set_ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

```python
# Set up the figure and axes
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Flatten the axes array for easier indexing
axes_flat = axes.flatten()

# Plot histograms for each numerical feature
for i, feature in enumerate(numerical_features):
    if i < 6:  # Only plot for the first 6 subplots
        sns.histplot(data=diamonds_df, x=feature, kde=True, ax=axes_flat[i], color='steelblue', bins=50)
        axes_flat[i].set_title(f'Distribusi {feature.capitalize()}')
        axes_flat[i].set_ylabel('Frekuensi')

plt.tight_layout()
plt.show()
```

```python
from ipywidgets import interact, widgets
import numpy as np

# Fungsi untuk menampilkan distribusi dengan interaktif bin width
def plot_histogram(feature, bin_width):
    plt.figure(figsize=(8, 6))
    bins = np.arange(min(df[feature]), max(df[feature]) + bin_width, bin_width)
    sns.histplot(df[feature], bins=bins, kde=True, color='darkcyan')
    plt.title(f'Distribution of {feature}', size=15)
    plt.xlabel(feature, size=12)
    plt.ylabel('Frequency', size=12)
    plt.show()

# Menampilkan slider interaktif untuk setiap fitur numerik
for feature in numerical_features:
    interact(plot_histogram, feature=widgets.fixed(feature), bin_width=(0.1, 1, 0.1))
```

```python
# Set up the figure and axes
fig, axes = plt.subplots(len(numerical_features), 1, figsize=(10, 20))

# Plot boxplots for each numerical feature
for i, feature in enumerate(numerical_features):
    sns.boxplot(x=diamonds_df[feature], ax=axes[i], color='lightseagreen')
    axes[i].set_title(f'Boxplot of {feature.capitalize()}')
    axes[i].set_xlabel(feature)

plt.tight_layout()
plt.show()
```
